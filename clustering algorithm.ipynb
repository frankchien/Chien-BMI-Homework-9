```python
"""
Frank Chien
BMI 500 Homework Week 9
Using the CSV obtained from the R script, this python script runs the output through K-mean clustering and outputs the result
This code is adapted from an earlier assignment (week 4) using K-means 
"""

import matplotlib.pyplot as plt
import pandas as pd
from sklearn import cluster, datasets

output_df = pd.read_csv("/Users/frankchien/Desktop/RishiHW/output.csv")
plt.scatter(output_df["lactate"],output_df["crp"])
plt.title("Lactate vs C-Reactive Protein in Patients with Impending AKI")
plt.xlabel("Lactate")
plt.ylabel("C-reactive Protein")
plt.show()
plt.savefig("scatter_uncolored.png",bbox_inches='tight')
plt.close('all')
```

```python
#start with 2 clusters
k_means_2 = cluster.KMeans(n_clusters=2)
sub_df = output_df[["lactate","crp"]]

#sub_df = breast_cancer_df[["mean radius","mean texture"]] #looks only at the first 2 attributes
target_2 = k_means_2.fit_predict(sub_df) #trains and predicts 2 clusters based on first 2 attributes

#adding 2-cluster predicted targets to database
output_df["predicted2"] = target_2

#plots the scatter plot with new targets based on kmeans2
plt.scatter(output_df["lactate"],output_df["crp"],c=output_df["predicted2"])
plt.title("Lactate vs C-Reactive Protein in Patients with Impending AKI")
plt.xlabel("Lactate")
plt.ylabel("C-reactive Protein")
plt.show()
plt.savefig("/Users/frankchien/Desktop/RishiHW/cluster.png")
plt.close('all')
```
